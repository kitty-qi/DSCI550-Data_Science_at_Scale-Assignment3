```python
# sample
df_en_sample = df_en.sample(n=2000, replace = False, random_state=1)
df_en_sample = df_en_sample.drop(['index'], axis=1)
df_en_sample.sort_values(by='Date', ascending = True, inplace = True)
df_en_sample = df_en_sample.reset_index(drop = True)
df_en_sample
```

```python
df_bn_sample = df_bn.sample(n=2000, replace = False, random_state=1)
df_bn_sample = df_bn_sample.drop(['index'], axis=1)
df_bn_sample.sort_values(by='Date', ascending = True, inplace = True)
df_bn_sample = df_bn_sample.reset_index(drop = True)
df_bn_sample
```

```python
df_it_sample = df_it.sample(n=3500, replace = False, random_state=2)
df_it_sample = df_it_sample.drop(['index'], axis=1)
df_it_sample.sort_values(by='Date', ascending = True, inplace = True)
df_it_sample = df_it_sample.reset_index(drop = True)
df_it_sample
```

```python
df_tl_sample = df_tl.reset_index(drop=True)
df_tl_sample = df_tl_sample.drop(['index'], axis=1)
df_tl_sample
```
