```python
import pandas as pd
import pysolr
import requests
from tqdm.notebook import tqdm
```

```python
df = pd.read_csv('combine_add_geo.csv')
```

```python
df
```

```python
df1 = df.sample(n = 1000, random_state=1)
```

```python
json_docs = []
doc_id = 0
for index, row in tqdm(df1.iterrows()):
    json_doc = {}
    json_doc['Gender'] = str(row['Gender'])
    json_doc['Age'] = str(row['Age'])
    json_doc['rtg_translate'] = str(row['rtg_translate'])
    json_doc['Detected Objects'] = str(row['Detected Objects'])
    
    
    json_doc["text"] = str(row['Gender']) + ' ' + str(row['Age']) + ' '+ str(row['Detected Objects'])+ ' ' + str(row['rtg_translate'])
    json_doc["id"] = str(doc_id)
    doc_id  = doc_id +1
    json_docs.append(json_doc)
```

```python
core_name = 'pixstory'
# Create a client instance. The timeout and authentication options are not required.
solr = pysolr.Solr('http://localhost:8983/solr/'+core_name, always_commit=True, timeout=10)
```

```python
for doc in tqdm(json_docs):
    # How you'd index data.
    solr.add([
        doc
    ])
```
